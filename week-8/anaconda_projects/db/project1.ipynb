```python
import tkinter as tk
import random

# Employee class
class Employee:
    # List of employees and tasks
    employees = [
        "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", "Andrew Ugwu",
        "Stella Mankinde", "Jane Akibo", "Ago James", "Michell Taiwo", "Abraham Jones", 
        "Nicole Anide", "Kosi Korso", "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
    ]
    tasks = [
        "Loading", "Transporting", "Reviewing Orders", "Customer Service", "Delivering Items"
    ]
    
    def __init__(self, name):
        self.name = name
        self.attendance = False
        self.task = None
    
    # Check if employee exists in the list
    def check_employee(self):
        if self.name in Employee.employees:
            return True
        else:
            return False
    
    # Mark attendance
    def take_attendance(self):
        self.attendance = True
    
    # Assign task randomly
    def assign_task(self):
        self.task = random.choice(Employee.tasks)
    
    # Refuse access if employee not found
    def refuse_access(self):
        return f"Access denied: {self.name} is not an employee."

# GUI Application
class EmployeeApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Employee Task Assignment System")
        
        # Create input field for employee name
        self.label = tk.Label(root, text="Enter your name:")
        self.label.pack()
        
        self.name_entry = tk.Entry(root)
        self.name_entry.pack()
        
        # Create buttons to check employee, take attendance, and assign task
        self.check_button = tk.Button(root, text="Check Employee", command=self.check_employee)
        self.check_button.pack()
        
        self.result_label = tk.Label(root, text="")
        self.result_label.pack()
        
    def check_employee(self):
        # Get employee name from input field
        employee_name = self.name_entry.get()
        
        # Create Employee object
        employee = Employee(employee_name)
        
        # Check if employee exists
        if employee.check_employee():
            employee.take_attendance()  # Mark attendance
            employee.assign_task()  # Assign task
            
            # Display result in the GUI
            self.result_label.config(text=f"Hello, {employee_name}!\nAttendance taken.\nYour task for today: {employee.task}")
        else:
            self.result_label.config(text=employee.refuse_access())  # Display refusal message

# Run the application
if __name__ == "__main__":
    root = tk.Tk()
    app = EmployeeApp(root)
    root.mainloop()
```
