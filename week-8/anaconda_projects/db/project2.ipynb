```python
import tkinter as tk

# DeliveryCharge class
class DeliveryCharge:
    def __init__(self, location, weight):
        self.location = location
        self.weight = weight
        self.charge = 0

    # Method to calculate the delivery charge based on location and weight
    def calculate_charge(self):
        if self.location == "PAU":
            if self.weight >= 10:
                self.charge = 2000
            else:
                self.charge = 1500
        elif self.location == "Epe":
            if self.weight >= 10:
                self.charge = 5000
            else:
                self.charge = 4000
        else:
            self.charge = "Invalid location"

# GUI Application
class DeliveryApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Delivery Charge Calculator")
        
        # Location input
        self.location_label = tk.Label(root, text="Enter delivery location (PAU or Epe):")
        self.location_label.pack()
        
        self.location_entry = tk.Entry(root)
        self.location_entry.pack()
        
        # Weight input
        self.weight_label = tk.Label(root, text="Enter package weight (kg):")
        self.weight_label.pack()
        
        self.weight_entry = tk.Entry(root)
        self.weight_entry.pack()

        # Button to calculate the charge
        self.calculate_button = tk.Button(root, text="Calculate Charge", command=self.calculate_charge)
        self.calculate_button.pack()
        
        # Label to display the result
        self.result_label = tk.Label(root, text="")
        self.result_label.pack()

    def calculate_charge(self):
        # Get user input
        location = self.location_entry.get().strip()
        try:
            weight = float(self.weight_entry.get().strip())
            
            # Create DeliveryCharge object
            delivery = DeliveryCharge(location, weight)
            delivery.calculate_charge()
            
            # Display the result
            if isinstance(delivery.charge, str):
                self.result_label.config(text=delivery.charge)
            else:
                self.result_label.config(text=f"Delivery Charge: N{delivery.charge}")
        except ValueError:
            self.result_label.config(text="Please enter a valid weight (numeric).")

# Run the application
if __name__ == "__main__":
    root = tk.Tk()
    app = DeliveryApp(root)
    root.mainloop()
```
